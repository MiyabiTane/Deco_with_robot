```python
class PersonalData:
  def __init__(self, date, number, age, sex, awareness, feeling_val, memorable):
    if (date == "1日目" and number == "1回目") or (date == "2日目" and number == "2回目") or (date == "3日目" and number == "1回目"):
      self.memories_talk = False
    elif (date == "1日目" and number == "2回目") or (date == "2日目" and number == "1回目") or (date == "3日目" and number == "2回目"):
      self.memories_talk = True
    self.age = age
    self.sex = sex
    self.awareness = awareness
    self.feeling_val = feeling_val
    self.memorable = memorable
  
  def show_data(self):
    print("===")
    print("思い出: {}, 年齢: {}, 性別: {}, 気づき: {}, 違和感: {}, 印象表情: {}".format(
            self.memories_talk, self.age, self.sex, self.awareness, self.feeling_val, self.memorable))
```

```python
def store_data(df):
  results = []
  for index, row in df.iterrows():
    result = PersonalData(date=row[0], number=row[1], age=row[2], sex=row[3], awareness=row[5], feeling_val=row[6], memorable=row[7])
    results.append(result)
  return results
```

```python
import matplotlib.pyplot as plt
import numpy as np 

def ana_sex_age(results):
  otona = []
  x = [i for i in range(81)]
  male_age = [0] * 81
  female_age = [0] * 81
  male_otona = 0
  male_kodomo = 0
  female_otona = 0
  female_kodomo = 0
  for result in results:
    if type(result.age) is float:
      if result.sex == "男":
        male_age[int(result.age)] += 1
        if result.age >= 18:
          male_otona += 1
        else:
          male_kodomo += 1
      if result.sex == "女":
        female_age[int(result.age)] += 1
        if result.age >= 18:
          female_otona += 1
        else:
          female_kodomo += 1
    else:
      otona.append((result.sex, result.age))
  # print(male_age)
  # print(female_age)
  print("Sum male: {}, female: {}".format(sum(male_age), sum(female_age)))
  print("male otona: {}, male_kodomo: {}, female_otona: {}, female_kodomo: {}".format(male_otona, male_kodomo, female_otona, female_kodomo))
  print("exception value: {}".format(otona))
  plt.figure(figsize=(10,4))
  plt.bar(x, male_age, color="blue") # alpha=0.5)
  plt.xlabel("age")
  plt.ylabel("the number of people")
  plt.savefig("age_distribution_male.png")
  plt.figure(figsize=(10,4))
  plt.bar(x, female_age, color = "orange") #alpha=0.5)
  plt.xlabel("age")
  plt.ylabel("the number of people")
  plt.savefig("age_distribution_female.png")
```
